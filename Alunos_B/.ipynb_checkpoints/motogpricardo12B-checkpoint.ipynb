```python
import random
import tabulate
import re
from IPython.display import clear_output
from time import sleep

fonte_pontos= open('Pontos_2022.txt','r',encoding='utf-8').read().splitlines()
fonte_pilotos= open('Pilotos_2022.txt','r',encoding='utf-8').read().splitlines()
fonte_pistas= open('Circuito_2022.txt','r',encoding='utf-8').read().splitlines()
cabecalho = ["Posição","Pontos","Piloto","Equipa","Mota"]   
todas_corridas=[]
for p in range(0,len(fonte_pilotos)):
    fonte_pilotos[p]=fonte_pilotos[p].split(',')
    fonte_pilotos[p][1]=fonte_pilotos[p][1]+ " " + fonte_pilotos[p][0]+" "+ fonte_pilotos[p][2]
    fonte_pilotos[p].remove(fonte_pilotos[p][0])
    fonte_pilotos[p].remove(fonte_pilotos[p][1])
for c in range (0,len(fonte_pontos)):
    fonte_pontos[c]=fonte_pontos[c].split(',')
    fonte_pontos[c].remove(fonte_pontos[c][0])
for c in range(len(fonte_pistas)):
    fonte_pistas[c]=fonte_pistas[c].split(',')    
for zeros in range(0, len(fonte_pilotos)-len(fonte_pontos)):
    fonte_pontos.append('0')  
    
def randomize(fonte_pilotos):        
    fonte_pilotos=random.sample(fonte_pilotos,len(fonte_pilotos)) 
    total = []
    for x, y in zip(fonte_pontos, fonte_pilotos):
        a = [x[0], y[0], y[1], y[2]]
        total.append(a)     
    print(tabulate.tabulate(total,headers=cabecalho,tablefmt='fancy_grid',stralign="center",numalign="center",showindex=range(1,len(fonte_pilotos)+1)))
    todas_corridas.append(total)
    
def campeonato():
    while True:
        print(tabulate.tabulate([["Selecionar Pista por Pista"],["Campeonato completo (20 Pistas)"]],tablefmt='fancy_grid',numalign="center",showindex=range(1,3)))
        selecao=input("Selecione o número pretendido: ")
        cleancode()
        if selecao and selecao.isnumeric() and int(selecao)==1:
            while True:
                print(tabulate.tabulate(fonte_pistas,headers=["Pista","País","Tammanho(metros)"],tablefmt='fancy_grid',stralign="center",numalign="center",showindex=range(1,len(fonte_pistas)+1)))
                selecao=input("Insira a sequência de numeros correspondentes aos circuitos pertendidos: ")
                if selecao:
                    selecao=re.split(r"\D+", selecao)
                    if all(selecao[x].isnumeric() and 0<int(selecao[x])<len(fonte_pistas)+1 for x in range(len(selecao))):
                        for y in selecao:
                            print("\n")
                            print(*fonte_pistas[int(y)-1], sep=", ", end=":\n")
                            randomize(fonte_pilotos)
                        print("\n"+"Resultados finais: ")    
                        resultados_totais("Pilotos")
                        resultados_totais("Equipas")    
                        input("Clique ENTER para prosseguir")
                        cleancode()
                        todas_corridas.clear()
                        break
                    else:
                        cleancode()
                        print("\n"+"ERRO!!! Por favor tente outra vez."+"\n")                                        
                else:
                    cleancode()
                    print("\n"+"ERRO!!! Por favor tente outra vez."+"\n")                    
        elif selecao and selecao.isnumeric() and int(selecao)==2:
            for m in range(len(fonte_pistas)):
                print("\n",end="")
                print(*fonte_pistas[m], sep=", ", end=":\n")
                randomize(fonte_pilotos)
            print("\n"+ "Resultados finais: ")
            resultados_totais("Pilotos")
            resultados_totais("Equipas")
            input("Clique ENTER para prosseguir")
            cleancode()
            todas_corridas.clear()            
        else:
            cleancode()
            print("\n"+"ERRO!!! Por favor tente outra vez."+"\n")

def resultados_totais(tipo_data):
    pilotos=[]
    if tipo_data=="Pilotos":
        data=1
    else:
        data=2  
    for corrida in range(len(todas_corridas)):
        resultado=todas_corridas[corrida]
        for piloto in range(len(resultado)):
            if any(pilotos[a][0]==resultado[piloto][data] for a in range(len(pilotos))):
                for a in range(len(pilotos)):
                    if pilotos[a][0] == resultado[piloto][data]:
                        pilotos[a][1] = int(pilotos[a][1]) + int(resultado[piloto][0])
            else:           
                pilotos.append([resultado[piloto][data],resultado[piloto][0]])
                
    pilotos.sort(reverse=True, key=lambda x: int(x[1]))
    print("\n"+tabulate.tabulate(pilotos,headers=[tipo_data,"Pontos"], tablefmt="fancy_grid", stralign="center", numalign="center", showindex=range(1,len(pilotos)+1)) + "\n")
                            
def cleancode():
    clear_output()
    sleep(0.02)                    
campeonato()
```
