```python
import random
import sys
from os import system

def Start():
    global pilotos
    global circuitos
    global pontos
    global gp_realizados
    global lista_menu
    pilotos = {}
    circuitos = {}
    pontos = {}
    gp_realizados = {"Último GP": []}
    lista_menu = ['- Pilotos', '- Grande Premio', '- Campeonato', '- Exit']

    with open("Pilotos_2022.txt", 'r', encoding="utf-8-sig") as fonte_pilotos:
        for x in fonte_pilotos:
            numero, nome, pais, equipa, mota = x.strip().split(',')
            pilotos[nome] = [numero, pais, equipa, mota, []]

    with open("Circuito_2022.txt", 'r', encoding="utf-8-sig") as fonte_circuitos:
        for c in fonte_circuitos:
            nome, pais, distancia = c.strip().split(',')
            circuitos[nome] = [pais, distancia, []]

    with open("Pontos_2022.txt", 'r', encoding="utf-8-sig") as fonte_pontos:
        for p in fonte_pontos:
            ponto = p.strip().split(',')
            pontos[ponto[0]] = ponto[1]

def Pilotos():
    lista_menu = ['- Listar pilotos', '- Classificação geral', '- Classificação por GP', '- Menu anterior', '- Exit']
    Menu(lista_menu, 0)

def ListarPilotos():
    print('\n', 'Nome'.rjust(10),'Número'.rjust(17), 'País'.rjust(9), 'Equipa'.rjust(17), 'Mota'.rjust(26))
    print('—'*85)
    for x in pilotos:
        print(x.ljust(24), pilotos[x][0].ljust(10), pilotos[x][1].ljust(8), pilotos[x][2].ljust(33), pilotos[x][3])
    print('\n', '\n')
    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 2)

def ClassificacaoGeral():
    res_final_gp = [(piloto, sum(pontuacao[1] for pontuacao in pilotos[piloto][4])) for piloto in pilotos]
    venc_ordenada = sorted(res_final_gp, key=lambda elem: elem[1], reverse=True)
    print('\nClassificação geral do Campeonato \n\nOrdem           Nome               Pontos')
    print('—'*41)

    for i, x in enumerate(venc_ordenada):
        print(' ', str(i+1).ljust(5), '  ', x[0].ljust(25), x[1])

    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 2)

def MenuClassificacaoPorGP():
    lista_menu = list(circuitos) + ['- Menu anterior', '- Exit']
    selection = Menu(lista_menu, 5)

    if selection == '- Menu anterior':
        system('cls')
        Pilotos()
    elif selection == '- Exit':
        Exit()
    else:
        ClassificacaoPorGP(selection)


def ClassificacaoPorGP(circuito):
    print('\n  Classificação do ' + circuito + '\n\nOrdem           Nome               Pontos')
    print('—'*41)
    resultados_ordenados = sorted(circuitos[circuito][2], key=lambda elem: elem[1], reverse=True)

    if resultados_ordenados:
        for x in resultados_ordenados:
            i = resultados_ordenados.index(x) + 1
            print(' ', str(i).ljust(5), '  ', x[0].ljust(25), x[1])

    else:
        print('Escolha um GP que já tenha sido realizado.')
    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 2)


def GrandePremio():
    lista_menu = ['- Realizar GP', '- Os vencedores por GP', '- Menu anterior', '- Exit']
    Menu(lista_menu, 0)


def RealizarGP():
    lista_menu = list(circuitos) + ['- Menu anterior', '- Exit']
    selection = Menu(lista_menu, 5)

    if selection == '- Menu anterior':
        system('cls')
        GrandePremio()
    elif selection == '- Exit':
        Exit()
    else:
        replica_pilotos = list(pilotos)
        classificacao_gp = []
        for i in range(1, len(pilotos)+1):
            pontuacao = int(pontos.get(str(i), 0))
            piloto = random.choice(replica_pilotos)
            replica_pilotos.remove(piloto)
            piloto_pontuacao = [piloto, pontuacao, pilotos[piloto][2]]
            pilotos[piloto][4].append([selection, pontuacao])
            classificacao_gp.append(piloto_pontuacao)

        circuitos[selection][2] = classificacao_gp
        gp_realizados[selection] = gp_realizados["Último GP"] = classificacao_gp

        print(f"\n\n  Grande Prémio de {selection} realizado com sucesso! \n\n" +
            "  A Classificação do GP realizado está disponível no menu anterior.\n")
        lista_menu = ['- Menu anterior', '- Exit']
        Menu(lista_menu, 3)


def VencedoresPorGP():
    lista_menu = list(gp_realizados) + ['- Menu anterior', '- Exit']

    selection = Menu(lista_menu, 5)

    if selection == "- Menu anterior":
        system('cls')
        GrandePremio()
    elif selection == "- Exit\n":
        Exit()
    else:
        resultados_ordenados = sorted(gp_realizados[selection], key=lambda elem: elem[1], reverse=True)

        print(f'\n  Classificação {selection} \n\nOrdem           Nome               Pontos')
        print('—'*41)
        for x in resultados_ordenados:
            i = resultados_ordenados.index(x) + 1
            print(' ', str(i).ljust(5), '  ', x[0].ljust(25), x[1])

        lista_menu = ['- Menu anterior', '- Exit']
        Menu(lista_menu, 3)


def Campeonato():
    lista_menu = ['- Realizar campeonato', '- Os dez primeiros do campeonato', '- Os primeiros de cada GP','- As três melhores equipas', '- Menu anterior', '- Exit']
    Menu(lista_menu, 0)


def RealizarCampeonato():
    Start()
    for gp in circuitos:
        replica_pilotos = list(pilotos)
        i = 1
        while replica_pilotos:
            try:
                pont = int(pontos[str(i)])
            except:
                pont = 0
            piloto = random.choice(replica_pilotos)
            replica_pilotos.remove(piloto)
            pilotos[piloto][4].append([gp, pont])
            circuitos[gp][2].append([piloto, pont, pilotos[piloto][2]])

            i += 1

    print('\n Campeonato realizado com sucesso! \n')

    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 4)


def DezPrimeiros():
    res_final_gp = [[x, sum(pontuacao[1] for pontuacao in pilotos[x][4])] for x in pilotos]
    tabela_ordenada = sorted(res_final_gp, key=lambda elem: elem[1], reverse=True)
    print('\n   Dez primeiros lugares do Campeonato  \n\nOrdem           Nome               Pontos')
    print('—'*41)
    for x in range(1, 11):
        piloto = tabela_ordenada[x-1][0]
        pontos = tabela_ordenada[x-1][1]
        print(' ', str(x).ljust(5), '  ', piloto.ljust(25), pontos)

    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 4)


def PrimeirosPorGP():
    print('\n   Primeiros Lugares por GP \n\n       Pista                                   Nome')
    print('—'*63)
    try:
        for circuito in circuitos:
            piloto = sorted(circuitos[circuito][2], key=lambda elem: elem[1], reverse=True)[0][0]
            print(circuito.ljust(40), piloto.ljust(22))
    except IndexError:
        print("O campeonato ainda não foi realizado!")

    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 4)


def MelhoresTresEquipas():
    pontos_equipas = {}
    for x in pilotos:
        soma = sum(pontuacao[1] for pontuacao in pilotos[x][4])
        equipa = pilotos[x][2]
        pontos_equipas[equipa] = pontos_equipas.get(equipa, 0) + soma

    pontos_ordenada = sorted(pontos_equipas, key=lambda elem: pontos_equipas[elem], reverse=True)

    print('\n   As três melhores equipas do Campeonato' + '\n\nOrdem            Equipa               Pontos')
    print('—'*47)

    for x in pontos_ordenada:
        i = pontos_ordenada.index(x) + 1
        print(' ', str(i).ljust(5), '  ', str(x).ljust(30), pontos_equipas[x])
        if i == 3:
            break

    lista_menu = ['- Menu anterior', '- Exit']
    Menu(lista_menu, 4)


def Exit():
    system('cls')
    print("Menu encerrado.")
    sys.exit()


def MenuAnterior():
    lista_menu = ['- Pilotos', '- Grande Premio', '- Campeonato', '- Exit']
    Menu(lista_menu, 0)


def display_menu(menu_items):
    for x in menu_items:
        print('\n', x, menu_items.get(x), '\n')


def Menu(lista_menu, flag):
    menu_items = dict(enumerate(lista_menu, start=1))
    while True:
        display_menu(menu_items)
        try:
            selection = int(input("Escolha um valor do menu: \n"))
        except:
            system('cls')
            print('Escolha um dos algarismos do menu.')
        else:
            try:
                selected = menu_items[selection]
                break
            except:
                system('cls')
                print('Escolha um dos algarismos do menu.')
    system('cls')

    nome_menu = {
        '- Pilotos': Pilotos,
        '- Grande Premio': GrandePremio,
        '- Campeonato': Campeonato,
        '- Exit': Exit,
        '- Os dez primeiros do campeonato': DezPrimeiros,
        '- As três melhores equipas': MelhoresTresEquipas,
        '- Menu anterior': MenuAnterior,
        '- Os primeiros de cada GP': PrimeirosPorGP,
        '- Realizar GP': RealizarGP,
        '- Classificação geral': ClassificacaoGeral,
        '- Realizar campeonato': RealizarCampeonato,
        '- Listar pilotos': ListarPilotos,
        '- Classificação por GP': MenuClassificacaoPorGP,
        '- Os vencedores por GP': VencedoresPorGP
                }
    if flag == 0:
        executa_funcao = nome_menu[selected]
        executa_funcao()
    elif flag == 1:
        RealizarGP()
    elif flag == 2:
        nome_menu['- Menu anterior'] = Pilotos
        executa_funcao = nome_menu[selected]
        executa_funcao()
    elif flag == 3:
        nome_menu['- Menu anterior'] = GrandePremio
        executa_funcao = nome_menu[selected]
        executa_funcao()
    elif flag == 4:
        nome_menu['- Menu anterior'] = Campeonato
        executa_funcao = nome_menu[selected]
        executa_funcao()
    elif flag == 5:
        return selected


system('cls')
Start()
Menu(lista_menu,0)
```
