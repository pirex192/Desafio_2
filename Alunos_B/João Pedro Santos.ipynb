```python
from tabulate import tabulate
import random
from IPython.display import clear_output
from time import sleep

f = open("Pilotos_2022.txt", "r", encoding='utf-8').read().splitlines()
g = open("Pontos_2022.txt", "r", encoding='utf-8').read().splitlines()
h= open("Circuito_2022.txt", "r", encoding='utf-8').read().splitlines()

def apagaroutput():
    clear_output()
    sleep(0.02)

#tabela circuitos
for x in range(0, len(h)):
        h[x] = h[x].split(",")
cabeçalho_pistas=["Pista", "País", "Distância (m)"]
h.append(["Correr campeonato"])

for x in range(0, len(f)):
    f[x] = f[x].split(",")
    f[x][0],f[x][1]=f[x][1],f[x][0]
cabeçalho_corrida=["Posição", "Piloto", "Número", "Pontos", "Nacionalidade", "Equipa", "Marca"]

for x in range(len(g)):
    g[x] = g[x].split(",")

for zeros in range(len(f)-len(g)):
    g.append([len(g)+1,'0'])

opções_menu=[["Fazer nova corrida"], ["Consultar classificações totais dos pilotos"], ["Consultar classificações totais das equipas"], ["Limpar memória"]]
menu=(tabulate(opções_menu, tablefmt='fancy_grid', stralign="left", numalign="center", showindex=range(1,len(opções_menu)+1)))
campeonato=[]
def resultadostotais(tipo):
    lista_pontos = []
    for total in campeonato:
        for x in total:
            if any(z[0] == x[tipo] for z in lista_pontos):
                for z in lista_pontos:
                    if z[0] == x[tipo]:
                        z[1] = int(z[1]) + int(x[3])
            else:
                lista_pontos.append([x[tipo],x[3]])
    lista_pontos.sort(reverse=True, key=lambda x: int(x[1]))
    return lista_pontos

def fazercorrida(c):
    fs = random.sample(f, len(f))
    
    total=[]
    for x, y in zip(g, fs):
        a = [x[0], y[0], y[1], x[1], y[2], y[3], y[4]]
        total.append(a)
    tabela_corrida=(tabulate(total, headers=(cabeçalho_corrida), tablefmt='fancy_grid', stralign="center", numalign="center"))

    hs=h[c-1]

    #menu


    print ("Pista: " + hs[0])
    print (tabela_corrida + "\n")
    return total

def menus(tipo, correr):
    while True:
        print (menu)
        selecao_menu=(input("Indique o número da opção que pretende: "))
        if selecao_menu and selecao_menu.isnumeric() and 0<int(selecao_menu)<5:
            apagaroutput()
            if int(selecao_menu) == 1:
                if tipo == "campeonato":
                    campeonato.clear()
                break
            elif int(selecao_menu) == 2:
                print (tabulate(resultadostotais(1), headers=(["Piloto","Pontos"]), tablefmt='fancy_grid', stralign="center", numalign="center") + "\n")
                input("Insira 1 para voltar")
                apagaroutput()
                if tipo == "corrida":
                    print((tabulate(correr, headers=(cabeçalho_corrida), tablefmt='fancy_grid', stralign="center", numalign="center")) + "\n")
                else:
                    for x in range(len(campeonato)):
                        print("Pista: " + h[x][0])
                        print((tabulate(campeonato[x], headers=(cabeçalho_corrida), tablefmt='fancy_grid', stralign="center", numalign="center")) + "\n")
            elif int(selecao_menu) == 3:
                print (tabulate(resultadostotais(5), headers=(["Equipa","Pontos"]), tablefmt='fancy_grid', stralign="center", numalign="center") + "\n")
                input("Insira 1 para voltar")
                apagaroutput()
                if tipo == "corrida":
                    print((tabulate(correr, headers=(cabeçalho_corrida), tablefmt='fancy_grid', stralign="center", numalign="center")) + "\n")
                else:
                    for x in range(len(campeonato)):
                        print("Pista: " + h[x][0])
                        print((tabulate(campeonato[x], headers=(cabeçalho_corrida), tablefmt='fancy_grid', stralign="center", numalign="center")) + "\n")
            else:
                campeonato.clear()
                break
        else:
            print("Insira um número correspondente a uma opção")      
            
while True:
    tabela_pistas=(tabulate(h, headers=(cabeçalho_pistas), tablefmt='fancy_grid', stralign="center", numalign="center", showindex=range(1,len(h)+1)))
    print (tabela_pistas)
    while True:
        pista=(input("Indique o número da pista: "))
        if pista and pista.isnumeric() and 0<int(pista)<len(h)+1:  
            apagaroutput()
            if int(pista)==len(h):
                campeonato.clear()
                for x in range(1, len(h)):
                    correr=fazercorrida(x)
                    campeonato.append(correr)
                menus("campeonato",correr)
                break
            elif 0<int(pista)<len(h):
                correr=fazercorrida(int(pista))
                campeonato.append(correr)
                menus("corrida", correr)
                break
            else:
                print("Erro")
        else:
            print("Insira um número correspondente a uma opção")
```
