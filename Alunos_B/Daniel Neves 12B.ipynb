```python
#Trabalho feito por Daniel Neves 12ºB
import re
import random
import tabulate
from tkinter import filedialog
from tkinter import Tk
from IPython.display import clear_output
from time import sleep

root = Tk()
root.withdraw()
root.wm_attributes('-topmost', 1)

def limparTela():
    clear_output()
    sleep(0.02)

def podio(nome, nome2, nome3):
    nomeLista = nome2.split()
    nomeLista2 = nome.split()
    nomeLista3 = nome3.split()

    h = 12

    x = ""
    for a in nomeLista:
        if len(a) > len(x):
            x = a
    cl = len(x)+4

    x = ""
    for a in nomeLista2:
        if len(a) > len(x):
            x = a
    cl2 = cl+len(x)+3

    x = ""
    for a in nomeLista3:
        if len(a) > len(x):
            x = a
    cl3 = cl2+len(x)+3
    
    x = len(nomeLista2)
    if len(nomeLista)+8 > x+12:
        x = len(nomeLista)-4
    if len(nomeLista3)+4 > x+12:
        x = len(nomeLista3)-8
    
    nomeL1 = []
    while len(nomeL1) < x+4-len(nomeLista):
        nomeL1.append("")
    for y in nomeLista:
        nomeL1.append(y)
    
    nomeL2 = []
    while len(nomeL2) < x-len(nomeLista2):
        nomeL2.append("")
    for y in nomeLista2:
        nomeL2.append(y)
    
    nomeL3 = []
    while len(nomeL3) < x+8-len(nomeLista3):
        nomeL3.append("")
    for y in nomeLista3:
        nomeL3.append(y)

    output = ""
    for k in range(h+x):
        for l in range(cl3):
            if 0 < l < cl-1 and k == x+4:
                output += "-"
            elif k == x and cl-len(nomeL1[k]) < l < cl2-len(nomeL1[k]):
                output += "-"
            elif cl2-1 < l < cl3-1 and k == x+8:
                output += "-"
        
            elif l == 0 and k > x+4 or l == cl//2 and k > x+5 and cl%2 != 0 or cl//2-2 < l < cl//2+1 and k > x+5 and cl%2 == 0:
                output += "|"
            elif l == cl2-1 and k > x+3 or l == cl-1 and k > x+3 or l == (cl2-cl)//2+cl-1 and k > x+3 and (cl2-cl)%2 == 0 or (cl2-cl)//2+cl-2 < l < (cl2-cl)//2+cl+1 and k > x+3 and (cl2-cl)%2 != 0:
                output += "|"
            elif x < k < x+4 and l == cl2-len(nomeL1[k]) or x < k < x+4 and l == cl-len(nomeL1[k]) or x+1 < k < x+4 and l == (cl2-cl)//2+cl-len(nomeL1[k]) and (cl2-cl)%2 == 0 or x+1 < k < x+4 and (cl2-cl)//2+cl-1-len(nomeL1[k]) < l < (cl2-cl)//2+cl+2-len(nomeL1[k]) and (cl2-cl)%2 != 0:
                output += "|"
            elif l == cl3-1 and k > x+8 or l == (cl3-cl2)//2+cl2-1 and k > x+9 and (cl3-cl2)%2 == 0 or (cl3-cl2)//2+cl2-2 < l < (cl3-cl2)//2+cl2+1 and k > x+9 and (cl3-cl2)%2 != 0:
                output += "|"
        
            elif l == cl-1 and k == x+4 or l == 0 and k == x+4 or l == cl//2 and k == x+5 and cl%2 != 0 or cl//2-2 < l < cl//2+1 and k == x+5 and cl%2 == 0:
                output += "@"
            elif k == x and l == cl2-len(nomeL1[k]) or k == x and l == cl-len(nomeL1[k]) or k == x+1 and l == (cl2-cl)//2+cl-len(nomeL1[k]) and (cl2-cl)%2 == 0 or k == x+1 and (cl2-cl)//2+cl-1-len(nomeL1[k]) < l < (cl2-cl)//2+cl+2-len(nomeL1[k]) and (cl2-cl)%2 != 0:
                output += "@"
            elif l == cl3-1 and k == x+8 or l == cl2-1 and k == x+8 or l == (cl3-cl2)//2+cl2-1 and k == x+9 and (cl3-cl2)%2 == 0 or (cl3-cl2)//2+cl2-2 < l < (cl3-cl2)//2+cl2+1 and k == x+9 and (cl3-cl2)%2 != 0:
                output += "@"
        
            elif k < x+4 and l == (cl-len(nomeL1[k]))//2:
                output += nomeL1[k]
            
            
            elif k < x and l == cl+(cl2-cl-len(nomeL2[k]))//2-len(nomeL1[k])+1:
                output += nomeL2[k]
            
            elif x+3 < k < x+8 and l == cl2+(cl3-cl2-len(nomeL3[k]))//2:
                output += nomeL3[k]
            elif x < k < x+4 and l == cl2+(cl3-cl2-len(nomeL3[k]))//2-len(nomeL1[k])+1:
                output += nomeL3[k]
            elif k < x and l == cl2+(cl3-cl2-len(nomeL3[k]))//2-len(nomeL2[k])-len(nomeL1[k])+2:
                output += nomeL3[k]
            elif k == x and l == cl2+(cl3-cl2-len(nomeL3[k]))//2-len(nomeL1[k])+1:
                output += nomeL3[k]
                
            elif k == h+x-1:
                output += "_"
            else:
                output += " "
        output += "\n"
    print(output)

try:
    while True:
        print("Selecione os pilotos.")
        pilotostxt = filedialog.askopenfilename(title="Selecionar Pilotos", filetypes=(("Ficheiros de Texto", "*.txt"),))
        if not pilotostxt:
            print("Ficheiro inválido. Tente novamente.")
        else:
            with open(pilotostxt, encoding='utf-8') as pilotos:
                pilotoslista = pilotos.readlines()
                if all(len(pilotoslista[piloto].strip().split(',')) == 5 for piloto in range(0, len(pilotoslista))):
                    break
                else:
                    print("Ficheiro inválido. Tente novamente.")

    while True:
        print("Selecione os circuitos.")
        circuitostxt = filedialog.askopenfilename(title="Selecionar Circuitos", filetypes=(("Ficheiros de Texto", "*.txt"),))
        if not circuitostxt:
            print("Ficheiro inválido. Tente novamente.")
        else:
            with open(circuitostxt, encoding='utf-8') as circuitos:
                circuitoslista = circuitos.readlines()
                if all(len(circuitoslista[circuito].strip().split(',')) == 3 for circuito in range(0, len(circuitoslista))):
                    break
                else:
                    print("Ficheiro inválido. Tente novamente.")

    while True:
        print("Selecione os pontos.")
        pontostxt = filedialog.askopenfilename(title="Selecionar Pontos", filetypes=(("Ficheiros de Texto", "*.txt"),))
        if not pontostxt:
            print("Ficheiro inválido. Tente novamente.")
        else:
            with open(pontostxt, encoding='utf-8') as pontos:
                pontoslista = pontos.readlines()
                if all(len(pontoslista[ponto].strip().split(',')) == 1 for ponto in range(0, len(pontoslista))):
                    break
                elif all(len(pontoslista[ponto].strip().split(',')) == 2 for ponto in range(0, len(pontoslista))):
                    break
                else:
                    print("Ficheiro inválido. Tente novamente.")
    root.destroy()
    
    with open(pilotostxt, encoding='utf-8') as pilotostxt:
        pilotoslista = pilotostxt.readlines()
    pilotos = []

    with open(circuitostxt, encoding='utf-8') as circuitostxt:
        circuitoslista = circuitostxt.readlines()
    circuitos = []
    
    with open(pontostxt, encoding='utf-8') as pontostxt:
        pontoslista = pontostxt.readlines()
    pontos = []
    
    for piloto in range(0, len(pilotoslista)):
        pilotos.append(pilotoslista[piloto].strip().split(","))
        pilotos[piloto][0], pilotos[piloto][1] = pilotos[piloto][1], pilotos[piloto][0]
        
    for circuito in range(0, len(circuitoslista)):
        circuitos.append(circuitoslista[circuito].strip().split(","))
        circuitos[circuito][2] = circuitos[circuito][2] + "m"
        
    for ponto in range(0, len(pontoslista)):
        pontos.append(pontoslista[ponto].strip().split(","))
        if len(pontos[ponto]) == 2:
            pontos[ponto].remove(pontos[ponto][0])

    for zeros in range(0, len(pilotos)-len(pontos)):
        pontos.append(['0'])

    print('\n')
    titulos = ["Pontos", "Piloto", "Número", "Nacionalidade", "Equipa", "Moto"]
    circuitoslista = circuitos.copy()
    circuitoslista.append(["Todos"])
    circuitoslista.append(["Combinação Personalizada"])
    
    campeonato = []
    def simular():
        resultado = list(map(lambda x, y:x + y, pontos, random.sample(pilotos, len(pilotos))))
        resultado.sort(reverse=True, key=lambda x: int(x[0]))
        campeonato.append(resultado)
        return resultado

    def menucampeonatos():
        print("Selecione o número correspondente à simulação pretendida:")
        print(tabulate.tabulate(circuitoslista, tablefmt="pretty", showindex=range(1, len(circuitoslista)+1)))
        circuitoSel = None
        while True:
            if circuitoSel and str(circuitoSel).isnumeric():
                print("Selecione o número correspondente à simulação pretendida:")
                print(tabulate.tabulate(circuitoslista, tablefmt="pretty", showindex=range(1, len(circuitoslista)+1)))
            circuitoSel = input("A sua seleção: ")
            if circuitoSel and circuitoSel.isnumeric() and 0 < int(circuitoSel) < len(circuitos)+1:
                limparTela()
                print(*circuitos[int(circuitoSel)-1], sep=(", "), end=(":\n\n"))
                resultado = simular()
                podio(resultado[0][1],resultado[1][1],resultado[2][1])
                escolharesultadosGP(resultado,circuitos[int(circuitoSel)-1])
                campeonato.clear()
            elif circuitoSel and circuitoSel.isnumeric() and int(circuitoSel) == len(circuitos)+1:
                limparTela()
                for x in range(len(circuitos)):
                    simular()
                a = 0
                while True:
                    if a == 0:
                        print(*circuitos[a], sep=(", "), end=(":\n\n"))
                        podio(campeonato[a][0][1],campeonato[a][1][1],campeonato[a][2][1])
                        a = a + escolharesultadosCampeonato(campeonato[a],0,a,a)
                    elif a == len(campeonato)-1:
                        print(*circuitos[a], sep=(", "), end=(":\n\n"))
                        podio(campeonato[a][0][1],campeonato[a][1][1],campeonato[a][2][1])
                        a = a + escolharesultadosCampeonato(campeonato[a],1,a,a)
                    elif a > len(campeonato)-1:
                        resultequipas = resultadostotais(4)
                        resultpilotos = resultadostotais(1)
                        podio(resultpilotos[0][0],resultpilotos[1][0],resultpilotos[2][0])
                        print("Selecione o número correspondente à opção pretendida:")
                        print(tabulate.tabulate([[1,"Ver resultados totais dos pilotos"],[2,"Ver resultados totais das equipas"],[3,"Fazer nova simulação"]], tablefmt="pretty"))
                        resultSel = None
                        while True:
                            if resultSel and str(resultSel).isnumeric():
                                podio(resultpilotos[0][0],resultpilotos[1][0],resultpilotos[2][0])
                                print("Selecione o número correspondente à opção pretendida:")
                                print(tabulate.tabulate([[1,"Ver resultados totais dos pilotos"],[2,"Ver resultados totais das equipas"],[3,"Fazer nova simulação"]], tablefmt="pretty"))
                            resultSel = input("A sua seleção: ")
                            if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                                limparTela()
                                print("\n" + tabulate.tabulate(resultpilotos, headers = ["Piloto","Pontos"], tablefmt="pretty") + "\n")
                                while True:
                                    resultSel = input("Selecione 1 para voltar: ")
                                    if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                                        limparTela()
                                        break
                                    else: print("Erro. Tente novamente.")
                            elif resultSel and resultSel.isnumeric() and int(resultSel) == 2:
                                limparTela()
                                print("\n" + tabulate.tabulate(resultequipas, headers = ["Equipa","Pontos"], tablefmt="pretty") + "\n")
                                while True:
                                    resultSel = input("Selecione 1 para voltar: ")
                                    if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                                        limparTela()
                                        break
                                    else: print("Erro. Tente novamente.")
                            elif resultSel and resultSel.isnumeric() and int(resultSel) == 3:
                                limparTela()
                                break
                            else:
                                print("Erro. Tente novamente.")
                        campeonato.clear()
                        break
                    else:
                        print(*circuitos[a], sep=(", "), end=(":\n\n"))
                        podio(campeonato[a][0][1],campeonato[a][1][1],campeonato[a][2][1])
                        a = a + escolharesultadosCampeonato(campeonato[a],2,a,a)
            elif circuitoSel and circuitoSel.isnumeric() and int(circuitoSel) == len(circuitos)+2:
                while True:
                    circuitoSel = input("Quais?: ")
                    if circuitoSel:
                        circuitosSel = re.split(r"\D+",circuitoSel)
                        circuitoSel = 1
                    if all(str(x).isnumeric() for x in circuitosSel):
                        circuitosSel.sort()
                        limparTela()
                        if circuitosSel:
                            for x in range(len(circuitosSel)):
                                simular()
                            a = 0
                            while True:
                                if a == 0:
                                    print(*circuitos[int(circuitosSel[a])-1], sep=(", "), end=(":\n\n"))
                                    podio(campeonato[a][0][1],campeonato[a][1][1],campeonato[a][2][1])
                                    a = a + escolharesultadosCampeonato(campeonato[a],0,int(circuitosSel[a])-1,a)
                                elif a == len(campeonato)-1:
                                    print(*circuitos[int(circuitosSel[a])-1], sep=(", "), end=(":\n\n"))
                                    podio(campeonato[a][0][1],campeonato[a][1][1],campeonato[a][2][1])
                                    a = a + escolharesultadosCampeonato(campeonato[a],1,int(circuitosSel[a])-1,a)
                                elif a > len(campeonato)-1:
                                    resultequipas = resultadostotais(4)
                                    resultpilotos = resultadostotais(1)
                                    podio(resultpilotos[0][0],resultpilotos[1][0],resultpilotos[2][0])
                                    print("Selecione o número correspondente à opção pretendida:")
                                    print(tabulate.tabulate([[1,"Ver resultados totais dos pilotos"],[2,"Ver resultados totais das equipas"],[3,"Fazer nova simulação"]], tablefmt="pretty"))
                                    resultSel = None
                                    while True:
                                        if resultSel and str(resultSel).isnumeric():
                                            podio(resultpilotos[0][0],resultpilotos[1][0],resultpilotos[2][0])
                                            print("Selecione o número correspondente à opção pretendida:")
                                            print(tabulate.tabulate([[1,"Ver resultados totais dos pilotos"],[2,"Ver resultados totais das equipas"],[3,"Fazer nova simulação"]], tablefmt="pretty"))
                                        resultSel = input("A sua seleção: ")
                                        if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                                            limparTela()
                                            print("\n" + tabulate.tabulate(resultpilotos, headers = ["Piloto","Pontos"], tablefmt="pretty") + "\n")
                                            while True:
                                                resultSel = input("Selecione 1 para voltar: ")
                                                if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                                                    limparTela()
                                                    break
                                                else: print("Erro. Tente novamente.")
                                        elif resultSel and resultSel.isnumeric() and int(resultSel) == 2:
                                            limparTela()
                                            print("\n" + tabulate.tabulate(resultequipas, headers = ["Equipa","Pontos"], tablefmt="pretty") + "\n")
                                            while True:
                                                resultSel = input("Selecione 1 para voltar: ")
                                                if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                                                    limparTela()
                                                    break
                                                else: print("Erro. Tente novamente.")
                                        elif resultSel and resultSel.isnumeric() and int(resultSel) == 3:
                                            limparTela()
                                            break
                                        else:
                                            print("Erro. Tente novamente.")
                                    campeonato.clear()
                                    break
                                else:
                                    print(*circuitos[int(circuitosSel[a])-1], sep=(", "), end=(":\n\n"))
                                    podio(campeonato[a][0][1],campeonato[a][1][1],campeonato[a][2][1])
                                    a = a + escolharesultadosCampeonato(campeonato[a],2,int(circuitosSel[a])-1,a)
                            break
                        else: print("Erro. Tente novamente.")
                    else: print("Erro. Tente novamente.")
            else: print("Erro. Tente novamente.")
    
    def escolharesultadosCampeonato(resultado,num,circuito,a):
        resultSel = None
        circuito = circuitos[circuito]
        print("Selecione o número correspondente à opção pretendida:")
        if num == 0:
            print(tabulate.tabulate([[1,"Ver resultados dos pilotos"],[2,"Ver resultados das equipas"],[3,"Próximo"],[4,"Último"]], tablefmt="pretty"))
        elif num == 1:
            print(tabulate.tabulate([[1,"Primeiro"],[2,"Anterior"],[3,"Ver resultados dos pilotos"],[4,"Ver resultados das equipas"],[5,"Ver resultados totais"]], tablefmt="pretty"))
        else:
            print(tabulate.tabulate([[1,"Anterior"],[2,"Ver resultados dos pilotos"],[3,"Ver resultados das equipas"],[4,"Próximo"],[5,"Último"]], tablefmt="pretty"))
        while True:
            if resultSel and str(resultSel).isnumeric():
                if num == 0 :
                    print(tabulate.tabulate([[1,"Ver resultados dos pilotos"],[2,"Ver resultados das equipas"],[3,"Próximo"],[4,"Último"]], tablefmt="pretty"))
                elif num == 1:
                    print(tabulate.tabulate([[1,"Primeiro"],[2,"Anterior"],[3,"Ver resultados dos pilotos"],[4,"Ver resultados das equipas"],[5,"Ver resultados totais"]], tablefmt="pretty"))
                else:
                    print(tabulate.tabulate([[1,"Anterior"],[2,"Ver resultados dos pilotos"],[3,"Ver resultados das equipas"],[4,"Próximo"],[5,"Último"]], tablefmt="pretty"))
            resultSel = input("A sua seleção: ")
            if resultSel and resultSel.isnumeric():
                if num == 0: resultSel = int(resultSel)+1
                if num == 1: resultSel = int(resultSel)-1
                if num == 1 and int(resultSel) == 4:
                    limparTela()
                    return len(campeonato)
                elif num == 1 and int(resultSel) == 0:
                    limparTela()
                    return -a
                elif num != 0 and int(resultSel) == 1:
                    limparTela()
                    return -1
                elif int(resultSel) == 2:
                    resultadospilotos(resultado,"Campeonato",circuito)
                elif int(resultSel) == 3:
                    resultadosequipas(resultado,"Campeonato",circuito)
                elif num != 1 and int(resultSel) == 4:
                    limparTela()
                    return 1
                elif num != 1 and int(resultSel) == 5:
                    limparTela()
                    return len(campeonato)-a-1
                else:
                    print("Erro. Tente novamente.")
            else:
                print("Erro. Tente novamente.")
        
    def escolharesultadosGP(resultado,circuito):
        print("Selecione o número correspondente à opção pretendida:")
        print(tabulate.tabulate([[1,"Fazer nova simulação"],[2,"Ver resultados dos pilotos"],[3,"Ver resultados das equipas"]], tablefmt="pretty"))
        while True:
            resultSel = input("A sua seleção: ")
            if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                campeonato.clear()
                limparTela()
                menucampeonatos()
            elif resultSel and resultSel.isnumeric() and int(resultSel) == 2:
                resultadospilotos(resultado,"GP",circuito)
            elif resultSel and resultSel.isnumeric() and int(resultSel) == 3:
                resultadosequipas(resultado,"GP",circuito)
            else: print("Erro. Tente novamente.")
    
    def resultadostotais(tipo):
        result = []
        for i in campeonato:
            for x in i:
                if any(z[0] == x[tipo] for z in result):
                    for z in result:
                        if z[0] == x[tipo]:
                            z[1] = int(z[1]) + int(x[0])
                else:
                    result.append([x[tipo],x[0]])
        result.sort(reverse=True, key=lambda x: int(x[1]))
        return result
    
    def resultadospilotos(resultado,tipo,circuito):
        limparTela()
        print(*circuito, sep=(", "), end=(":\n\n"))
        print(tabulate.tabulate(resultado, headers = titulos, tablefmt="pretty", showindex=range(1, len(resultado)+1)) + '\n')
        while True:
            resultSel = input("Selecione 1 para voltar: ")
            if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                break
            else: print("Erro. Tente novamente.")
        limparTela()
        print(*circuito, sep=(", "), end=(":\n\n"))
        podio(resultado[0][1],resultado[1][1],resultado[2][1])
        if tipo == "GP":
            escolharesultadosGP(resultado,circuito)
    
    def resultadosequipas(resultado,tipo,circuito):
        limparTela()
        print(*circuito, sep=(", "), end=(":\n\n"))
        resultequipas = []
        for x in resultado:
            if any(z[0] == x[4] for z in resultequipas):
                for z in resultequipas:
                    if z[0] == x[4]:
                        z[1] = int(z[1]) + int(x[0])
            else:
                resultequipas.append([x[4],x[0]])
        resultequipas.sort(reverse=True, key=lambda x: int(x[1]))
        print("\n" + tabulate.tabulate(resultequipas, headers = ["Equipa","Pontos"], tablefmt="pretty") + "\n")
        while True:
            resultSel = input("Selecione 1 para voltar: ")
            if resultSel and resultSel.isnumeric() and int(resultSel) == 1:
                break
            else: print("Erro. Tente novamente.")
        limparTela()
        print(*circuito, sep=(", "), end=(":\n\n"))
        podio(resultado[0][1],resultado[1][1],resultado[2][1])
        if tipo == "GP":
            escolharesultadosGP(resultado,circuito)
                    
    limparTela()
    menucampeonatos()

except KeyboardInterrupt:
    print("Programa interrompido com sucesso.")
```
