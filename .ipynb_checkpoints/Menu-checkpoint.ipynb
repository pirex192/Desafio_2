```python
"""
1. Add a menu to a console application to manage activities.
2. Run a selected function.
3. Clear the output
4. Display the menu again or exit if done is selected
"""

import sys
from os import system

fonte_pilotos = open("../../jupyter/Desafio_2/Pilotos_2022.txt",'r')
fonte_circuitos = open("../../jupyter/Desafio_2/Circuito_2022.txt",'r')
fonte_pontos = open("../../jupyter/Desafio_2/Pontos_2022.txt",'r')

pilotos = {}
circuitos = {}
pontos = {}

for x in fonte_pilotos:
    piloto = x.strip().split(',')
    pilotos[piloto[1]] = [piloto[0], piloto[2], piloto[3], piloto[4]]
fonte_pilotos.close()

for c in fonte_circuitos:
    circuito = c.strip().split(',')
    circuitos[circuito[0]] = [circuito[1], circuito[2]]
fonte_circuitos.close()

for p in fonte_pontos:
    ponto = p.strip().split(',')
    pontos[ponto[0]] = ponto[1]
fonte_pontos.close()

def display_menu(menu_items):
    print(menu_items)

def Pilotos():
    print("Escolha uma das seguintes opções") # Simulate function output.
    lista_menu = ['Listar pilotos', 'Classificação geral', 'Classificação por GP', 'Menu anterior', 'Exit']    
    Menu(lista_menu)    
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout


def GrandePremio():
    print("Escolha uma das seguintes opções") # Simulate function output.
    lista_menu = ['Escolher circuito', 'Os vencedores por GP', 'Menu anterior', 'Exit']    
    Menu(lista_menu)
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout


def Campeonato():
    print("Escolha uma das seguintes opções")
    lista_menu = ['Realizar campeonato', 'Os dez primeiros do campeonato', 'Os primeiros de cada GP','As três melhores equipas', 'Menu anterior', 'Exit']    
    Menu(lista_menu)
    input("Press Enter to Continue\n")
    system('cls')


def Exit():
    name = 'Piloto'    
    system('cls')  # clears stdout
    print("Goodbye")
    sys.exit()

def ListarPilotos():
    lista_menu = ['Menu anterior', 'Exit']    
    Menu(lista_menu)
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
    
def EscolherCircuito():
    lista_menu = ['Realizar GP', 'Classificação geral', 'Poole position', 'Menu anterior', 'Exit']    
    Menu(lista_menu)
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout

def RealizarCampeonato():
    print('Realizar campeonato') # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
    
def PoolePosition():
    print('Poole position') # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
    
def ClassificaçãoGeral():
    print('Classificação geral') # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
    
def RealizarGP():
    print('Realizar GP') # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
    
def PrimeirosCadaGP():
    print('Os primeiros de cada GP') # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout

def DezPrimeiros():
    print("DezPrimeiros") # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
    
def TresEquipas():
    print("TresEquipas") # Simulate function output.
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout

def MenuAnterior():
    lista_menu = ['Pilotos', 'Grande Premio', 'Campeonato', 'Exit']
    Menu(lista_menu)
    input("Press Enter to Continue\n")
    system('cls')  # clears stdout
        
#functions_names = [Pilotos, GrandePremio, Campeonato, Exit]
lista_menu = ['Pilotos', 'Grande Premio', 'Campeonato', 'Exit']

def Menu(lista_menu):
    # Create a menu dictionary where the key is an integer number and the
    # value is a function name.
#    menu_items = dict(enumerate(functions_names, start=1))
    menu_items = dict(enumerate(lista_menu, start=1))
#    print(menu_items)    
#    while True:
#        print(menu_items)
    display_menu(menu_items)
    selection = int(input("Please enter your selection number: "))  # Get function key
    selected = menu_items[selection]  # Gets the function name
    nome_menu = {'Pilotos' : Pilotos,
                 'Grande Premio' : GrandePremio,
                 'Campeonato' : Campeonato,
                 'Exit' : Exit,
                 'Os dez primeiros' : DezPrimeiros,
                 'As três melhores equipas' : TresEquipas,
                 'Menu anterior' : MenuAnterior,
                 'Os primeiros de cada GP' : PrimeirosCadaGP,
                 'Realizar GP' : RealizarGP,
                 'Classificação geral' : ClassificaçãoGeral,
                 'Poole position' : PoolePosition,
                 'Realizar campeonato' : RealizarCampeonato,
                 'Escolher circuito' : EscolherCircuito,
                 'Listar pilotos' : ListarPilotos
    }
    executa_funcao = nome_menu[selected]
#    print(selected_value)
    executa_funcao()  # add parentheses to call the function


#if __name__ == "__main__":
Menu(lista_menu)
```
