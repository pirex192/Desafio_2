```python
#Estrutura de dados gerada com o motoGP:
#circuitos2022 = {'circuito': ['país', 'distância', [('equipa', pontos, 'piloto')]]}

circuitos2022 = {'Circuit Ricardo Tormo': ['Espanha', '4005', [('Team SUZUKI ECSTAR', 25, 'Alex Rins'), ('Aprilia Racing', 20, 'Maverick Vitalis'), ('Aprilia Racing', 16, 'Lorenzo Savadori'), ('WithU Yamaha RNF MotoGP Team', 13, 'Cal Crutchlow'), ('Mooney VR46 Racing Team', 11, 'Marco Bezzecchi'), ('Repsol Honda Team', 10, 'Pol Espargaro'), ('LCR Honda CASTROL', 9, 'Alex Marquez'), ('Team SUZUKI ECSTAR', 8, 'Joan Mir'), ('Repsol Honda Team', 7, 'Stefan Bradl'), ('Monster Energy Yamaha MotoGP', 6, 'Franco Morbidelli'), ('Gresini Racing MotoGP', 5, 'Fabio Di Giannantonio'), ('Team SUZUKI ECSTAR', 4, 'Danilo Petrucci'), ('Repsol Honda Team', 3, 'Marc Marquez'), ('Gresini Racing MotoGP', 2, 'Enea Bastianini'), ('LCR Honda IDEMITSU', 1, 'Tetsuta Nagashima'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('Ducati Lenovo Team', 0, 'Jack Miller'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder')]], 'Sepang International Circuit': ['Malasia', '5543', [('WithU Yamaha RNF MotoGP Team', 25, 'Andrea Dovizioso'), ('Mooney VR46 Racing Team', 20, 'Marco Bezzecchi'), ('Prima Pramac Racing', 16, 'Johann Zarco'), ('Red Bull KTM Factory Racing', 13, 'Miguel Oliveira'), ('Ducati Lenovo Team', 11, 'Francesco Bagnaia'), ('Monster Energy Yamaha MotoGP', 10, 'Franco Morbidelli'), ('Repsol Honda Team', 9, 'Marc Marquez'), ('Aprilia Racing', 8, 'Aleix Espargaro'), ('Team SUZUKI ECSTAR', 7, 'Kazuki Watanabe'), ('Prima Pramac Racing', 6, 'Jorge Martin'), ('Mooney VR46 Racing Team', 5, 'Luca Marini'), ('Tech3 KTM Factory Racing', 4, 'Raul Fernandez'), ('WithU Yamaha RNF MotoGP Team', 3, 'Cal Crutchlow'), ('Repsol Honda Team', 2, 'Stefan Bradl'), ('Aruba.it Racing', 1, 'Michele Pirro'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Aprilia Racing', 0, 'Lorenzo Savadori'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('Ducati Lenovo Team', 0, 'Jack Miller')]], 'Phillip Island': ['Australia', '4448', [('Gresini Racing MotoGP', 25, 'Enea Bastianini'), ('LCR Honda IDEMITSU', 20, 'Tetsuta Nagashima'), ('Mooney VR46 Racing Team', 16, 'Marco Bezzecchi'), ('Tech3 KTM Factory Racing', 13, 'Raul Fernandez'), ('Team SUZUKI ECSTAR', 11, 'Alex Rins'), ('Team SUZUKI ECSTAR', 10, 'Joan Mir'), ('Monster Energy Yamaha MotoGP', 9, 'Fabio Quartararo'), ('Prima Pramac Racing', 8, 'Johann Zarco'), ('Red Bull KTM Factory Racing', 7, 'Brad Binder'), ('Ducati Lenovo Team', 6, 'Jack Miller'), ('Tech3 KTM Factory Racing', 5, 'Remy Gardner'), ('Repsol Honda Team', 4, 'Stefan Bradl'), ('Aprilia Racing', 3, 'Maverick Vitalis'), ('Aprilia Racing', 2, 'Lorenzo Savadori'), ('LCR Honda CASTROL', 1, 'Alex Marquez'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow')]], 'Chang International Circuit': ['Tailandia', '4554', [('WithU Yamaha RNF MotoGP Team', 25, 'Darryn Binder'), ('WithU Yamaha RNF MotoGP Team', 20, 'Andrea Dovizioso'), ('Mooney VR46 Racing Team', 16, 'Marco Bezzecchi'), ('Red Bull KTM Factory Racing', 13, 'Miguel Oliveira'), ('Prima Pramac Racing', 11, 'Johann Zarco'), ('Ducati Lenovo Team', 10, 'Jack Miller'), ('Repsol Honda Team', 9, 'Stefan Bradl'), ('WithU Yamaha RNF MotoGP Team', 8, 'Cal Crutchlow'), ('Aruba.it Racing', 7, 'Michele Pirro'), ('Monster Energy Yamaha MotoGP', 6, 'Fabio Quartararo'), ('Aprilia Racing', 5, 'Lorenzo Savadori'), ('Gresini Racing MotoGP', 4, 'Enea Bastianini'), ('Aprilia Racing', 3, 'Aleix Espargaro'), ('Repsol Honda Team', 2, 'Marc Marquez'), ('Prima Pramac Racing', 1, 'Jorge Martin'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci')]], 'Mobility Resort Motegi': ['Japao', '4801', [('Ducati Lenovo Team', 25, 'Francesco Bagnaia'), ('Gresini Racing MotoGP', 20, 'Enea Bastianini'), ('Aprilia Racing', 16, 'Aleix Espargaro'), ('Aprilia Racing', 13, 'Lorenzo Savadori'), ('Monster Energy Yamaha MotoGP', 11, 'Franco Morbidelli'), ('Monster Energy Yamaha MotoGP', 10, 'Fabio Quartararo'), ('Prima Pramac Racing', 9, 'Johann Zarco'), ('Ducati Lenovo Team', 8, 'Jack Miller'), ('WithU Yamaha RNF MotoGP Team', 7, 'Cal Crutchlow'), ('LCR Honda IDEMITSU', 6, 'Takaaki Nakagami'), ('Team SUZUKI ECSTAR', 5, 'Danilo Petrucci'), ('WithU Yamaha RNF MotoGP Team', 4, 'Andrea Dovizioso'), ('Team SUZUKI ECSTAR', 3, 'Kazuki Watanabe'), ('Mooney VR46 Racing Team', 2, 'Marco Bezzecchi'), ('LCR Honda CASTROL', 1, 'Alex Marquez'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Repsol Honda Team', 0, 'Stefan Bradl')]], 'MotorLand Aragon': ['Espanha', '5077', [('Repsol Honda Team', 25, 'Marc Marquez'), ('WithU Yamaha RNF MotoGP Team', 20, 'Darryn Binder'), ('Aprilia Racing', 16, 'Aleix Espargaro'), ('Tech3 KTM Factory Racing', 13, 'Raul Fernandez'), ('Prima Pramac Racing', 11, 'Jorge Martin'), ('LCR Honda IDEMITSU', 10, 'Tetsuta Nagashima'), ('Team SUZUKI ECSTAR', 9, 'Kazuki Watanabe'), ('LCR Honda IDEMITSU', 8, 'Takaaki Nakagami'), ('Repsol Honda Team', 7, 'Stefan Bradl'), ('Red Bull KTM Factory Racing', 6, 'Brad Binder'), ('Team SUZUKI ECSTAR', 5, 'Alex Rins'), ('Team SUZUKI ECSTAR', 4, 'Joan Mir'), ('Team SUZUKI ECSTAR', 3, 'Danilo Petrucci'), ('Tech3 KTM Factory Racing', 2, 'Remy Gardner'), ('Aprilia Racing', 1, 'Lorenzo Savadori'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Ducati Lenovo Team', 0, 'Jack Miller'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini')]], 'Misano World Circuit Marco Simoncelli': ['San Marino', '4226', [('WithU Yamaha RNF MotoGP Team', 25, 'Andrea Dovizioso'), ('Prima Pramac Racing', 20, 'Jorge Martin'), ('Ducati Lenovo Team', 16, 'Jack Miller'), ('LCR Honda IDEMITSU', 13, 'Takaaki Nakagami'), ('Repsol Honda Team', 11, 'Pol Espargaro'), ('Gresini Racing MotoGP', 10, 'Fabio Di Giannantonio'), ('Red Bull KTM Factory Racing', 9, 'Brad Binder'), ('Mooney VR46 Racing Team', 8, 'Marco Bezzecchi'), ('Mooney VR46 Racing Team', 7, 'Luca Marini'), ('Monster Energy Yamaha MotoGP', 6, 'Fabio Quartararo'), ('Ducati Lenovo Team', 5, 'Francesco Bagnaia'), ('Aprilia Racing', 4, 'Aleix Espargaro'), ('Prima Pramac Racing', 3, 'Johann Zarco'), ('Repsol Honda Team', 2, 'Stefan Bradl'), ('Aruba.it Racing', 1, 'Michele Pirro'), ('Aprilia Racing', 0, 'Lorenzo Savadori'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe')]], 'Red Bull Ring - Spielberg': ['Austria', '4348', [('Gresini Racing MotoGP', 25, 'Fabio Di Giannantonio'), ('Team SUZUKI ECSTAR', 20, 'Alex Rins'), ('Tech3 KTM Factory Racing', 16, 'Remy Gardner'), ('Gresini Racing MotoGP', 13, 'Enea Bastianini'), ('LCR Honda CASTROL', 11, 'Alex Marquez'), ('WithU Yamaha RNF MotoGP Team', 10, 'Andrea Dovizioso'), ('Aruba.it Racing', 9, 'Michele Pirro'), ('Prima Pramac Racing', 8, 'Jorge Martin'), ('Repsol Honda Team', 7, 'Pol Espargaro'), ('WithU Yamaha RNF MotoGP Team', 6, 'Darryn Binder'), ('Team SUZUKI ECSTAR', 5, 'Kazuki Watanabe'), ('Monster Energy Yamaha MotoGP', 4, 'Franco Morbidelli'), ('Aprilia Racing', 3, 'Maverick Vitalis'), ('Red Bull KTM Factory Racing', 2, 'Brad Binder'), ('Ducati Lenovo Team', 1, 'Jack Miller'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Aprilia Racing', 0, 'Lorenzo Savadori'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi')]], 'Silverstone Circuit': ['Reino Unido', '5900', [('Mooney VR46 Racing Team', 25, 'Marco Bezzecchi'), ('Aprilia Racing', 20, 'Maverick Vitalis'), ('Gresini Racing MotoGP', 16, 'Fabio Di Giannantonio'), ('Team SUZUKI ECSTAR', 13, 'Alex Rins'), ('Monster Energy Yamaha MotoGP', 11, 'Fabio Quartararo'), ('Team SUZUKI ECSTAR', 10, 'Kazuki Watanabe'), ('Gresini Racing MotoGP', 9, 'Enea Bastianini'), ('WithU Yamaha RNF MotoGP Team', 8, 'Andrea Dovizioso'), ('Repsol Honda Team', 7, 'Stefan Bradl'), ('Mooney VR46 Racing Team', 6, 'Luca Marini'), ('Prima Pramac Racing', 5, 'Jorge Martin'), ('Repsol Honda Team', 4, 'Marc Marquez'), ('LCR Honda IDEMITSU', 3, 'Tetsuta Nagashima'), ('Monster Energy Yamaha MotoGP', 2, 'Franco Morbidelli'), ('Tech3 KTM Factory Racing', 1, 'Raul Fernandez'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Aprilia Racing', 0, 'Lorenzo Savadori'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Ducati Lenovo Team', 0, 'Jack Miller'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci')]], 'TT Circuit Assen': ['Holanda', '4542', [('Team SUZUKI ECSTAR', 25, 'Alex Rins'), ('Prima Pramac Racing', 20, 'Jorge Martin'), ('Aruba.it Racing', 16, 'Michele Pirro'), ('Mooney VR46 Racing Team', 13, 'Luca Marini'), ('LCR Honda IDEMITSU', 11, 'Takaaki Nakagami'), ('Tech3 KTM Factory Racing', 10, 'Remy Gardner'), ('Gresini Racing MotoGP', 9, 'Enea Bastianini'), ('Red Bull KTM Factory Racing', 8, 'Brad Binder'), ('Mooney VR46 Racing Team', 7, 'Marco Bezzecchi'), ('Aprilia Racing', 6, 'Maverick Vitalis'), ('Aprilia Racing', 5, 'Lorenzo Savadori'), ('Tech3 KTM Factory Racing', 4, 'Raul Fernandez'), ('WithU Yamaha RNF MotoGP Team', 3, 'Darryn Binder'), ('Monster Energy Yamaha MotoGP', 2, 'Fabio Quartararo'), ('Repsol Honda Team', 1, 'Marc Marquez'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Ducati Lenovo Team', 0, 'Jack Miller'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('LCR Honda CASTROL', 0, 'Alex Marquez')]], 'Sachsenring': ['Alemanha', '3671', [('Monster Energy Yamaha MotoGP', 25, 'Franco Morbidelli'), ('Ducati Lenovo Team', 20, 'Jack Miller'), ('Team SUZUKI ECSTAR', 16, 'Kazuki Watanabe'), ('Prima Pramac Racing', 13, 'Jorge Martin'), ('Team SUZUKI ECSTAR', 11, 'Alex Rins'), ('LCR Honda CASTROL', 10, 'Alex Marquez'), ('Tech3 KTM Factory Racing', 9, 'Remy Gardner'), ('WithU Yamaha RNF MotoGP Team', 8, 'Darryn Binder'), ('Aprilia Racing', 7, 'Maverick Vitalis'), ('Gresini Racing MotoGP', 6, 'Fabio Di Giannantonio'), ('Mooney VR46 Racing Team', 5, 'Luca Marini'), ('LCR Honda IDEMITSU', 4, 'Takaaki Nakagami'), ('Tech3 KTM Factory Racing', 3, 'Raul Fernandez'), ('WithU Yamaha RNF MotoGP Team', 2, 'Cal Crutchlow'), ('Team SUZUKI ECSTAR', 1, 'Danilo Petrucci'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi'), ('Aprilia Racing', 0, 'Lorenzo Savadori')]], 'Circuit de Barcelona-Catalunya': ['Espanha', '4657', [('Team SUZUKI ECSTAR', 25, 'Alex Rins'), ('Gresini Racing MotoGP', 20, 'Enea Bastianini'), ('Team SUZUKI ECSTAR', 16, 'Danilo Petrucci'), ('Repsol Honda Team', 13, 'Marc Marquez'), ('Monster Energy Yamaha MotoGP', 11, 'Fabio Quartararo'), ('Aprilia Racing', 10, 'Maverick Vitalis'), ('Gresini Racing MotoGP', 9, 'Fabio Di Giannantonio'), ('WithU Yamaha RNF MotoGP Team', 8, 'Darryn Binder'), ('Ducati Lenovo Team', 7, 'Jack Miller'), ('Mooney VR46 Racing Team', 6, 'Luca Marini'), ('Repsol Honda Team', 5, 'Stefan Bradl'), ('Tech3 KTM Factory Racing', 4, 'Raul Fernandez'), ('Red Bull KTM Factory Racing', 3, 'Brad Binder'), ('Prima Pramac Racing', 2, 'Johann Zarco'), ('Ducati Lenovo Team', 1, 'Francesco Bagnaia'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Aprilia Racing', 0, 'Lorenzo Savadori')]], 'Autodromo Internazionale del Mugello': ['Italia', '5245', [('Monster Energy Yamaha MotoGP', 25, 'Fabio Quartararo'), ('Team SUZUKI ECSTAR', 20, 'Danilo Petrucci'), ('Team SUZUKI ECSTAR', 16, 'Alex Rins'), ('WithU Yamaha RNF MotoGP Team', 13, 'Darryn Binder'), ('Tech3 KTM Factory Racing', 11, 'Remy Gardner'), ('Aprilia Racing', 10, 'Lorenzo Savadori'), ('Aruba.it Racing', 9, 'Michele Pirro'), ('Tech3 KTM Factory Racing', 8, 'Raul Fernandez'), ('Aprilia Racing', 7, 'Aleix Espargaro'), ('Mooney VR46 Racing Team', 6, 'Marco Bezzecchi'), ('Ducati Lenovo Team', 5, 'Jack Miller'), ('Team SUZUKI ECSTAR', 4, 'Joan Mir'), ('LCR Honda IDEMITSU', 3, 'Takaaki Nakagami'), ('Repsol Honda Team', 2, 'Marc Marquez'), ('Prima Pramac Racing', 1, 'Jorge Martin'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('LCR Honda CASTROL', 0, 'Alex Marquez')]], 'Le Mans': ['Franca', '4185', [('Mooney VR46 Racing Team', 25, 'Marco Bezzecchi'), ('Gresini Racing MotoGP', 20, 'Fabio Di Giannantonio'), ('Aprilia Racing', 16, 'Lorenzo Savadori'), ('Repsol Honda Team', 13, 'Stefan Bradl'), ('Ducati Lenovo Team', 11, 'Jack Miller'), ('Aprilia Racing', 10, 'Maverick Vitalis'), ('Team SUZUKI ECSTAR', 9, 'Kazuki Watanabe'), ('Red Bull KTM Factory Racing', 8, 'Brad Binder'), ('Mooney VR46 Racing Team', 7, 'Luca Marini'), ('Red Bull KTM Factory Racing', 6, 'Miguel Oliveira'), ('Team SUZUKI ECSTAR', 5, 'Danilo Petrucci'), ('Ducati Lenovo Team', 4, 'Francesco Bagnaia'), ('Gresini Racing MotoGP', 3, 'Enea Bastianini'), ('LCR Honda IDEMITSU', 2, 'Tetsuta Nagashima'), ('WithU Yamaha RNF MotoGP Team', 1, 'Cal Crutchlow'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder')]], 'Circuito de Jerez - Angel Nieto': ['Espanha', '4423', [('Team SUZUKI ECSTAR', 25, 'Kazuki Watanabe'), ('WithU Yamaha RNF MotoGP Team', 20, 'Cal Crutchlow'), ('Prima Pramac Racing', 16, 'Johann Zarco'), ('WithU Yamaha RNF MotoGP Team', 13, 'Darryn Binder'), ('Ducati Lenovo Team', 11, 'Francesco Bagnaia'), ('Mooney VR46 Racing Team', 10, 'Marco Bezzecchi'), ('Repsol Honda Team', 9, 'Marc Marquez'), ('LCR Honda IDEMITSU', 8, 'Tetsuta Nagashima'), ('Gresini Racing MotoGP', 7, 'Fabio Di Giannantonio'), ('LCR Honda IDEMITSU', 6, 'Takaaki Nakagami'), ('WithU Yamaha RNF MotoGP Team', 5, 'Andrea Dovizioso'), ('Mooney VR46 Racing Team', 4, 'Luca Marini'), ('Red Bull KTM Factory Racing', 3, 'Brad Binder'), ('Aruba.it Racing', 2, 'Michele Pirro'), ('Aprilia Racing', 1, 'Maverick Vitalis'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Ducati Lenovo Team', 0, 'Jack Miller'), ('Aprilia Racing', 0, 'Lorenzo Savadori'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini')]], 'Autodromo Internacional do Algarve': ['Portugal', '4592', [('Team SUZUKI ECSTAR', 25, 'Danilo Petrucci'), ('Monster Energy Yamaha MotoGP', 20, 'Franco Morbidelli'), ('Monster Energy Yamaha MotoGP', 16, 'Fabio Quartararo'), ('Aruba.it Racing', 13, 'Michele Pirro'), ('LCR Honda CASTROL', 11, 'Alex Marquez'), ('Mooney VR46 Racing Team', 10, 'Luca Marini'), ('Ducati Lenovo Team', 9, 'Jack Miller'), ('WithU Yamaha RNF MotoGP Team', 8, 'Cal Crutchlow'), ('Team SUZUKI ECSTAR', 7, 'Alex Rins'), ('Mooney VR46 Racing Team', 6, 'Marco Bezzecchi'), ('Tech3 KTM Factory Racing', 5, 'Remy Gardner'), ('Red Bull KTM Factory Racing', 4, 'Miguel Oliveira'), ('Aprilia Racing', 3, 'Lorenzo Savadori'), ('Prima Pramac Racing', 2, 'Jorge Martin'), ('Ducati Lenovo Team', 1, 'Francesco Bagnaia'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Repsol Honda Team', 0, 'Pol Espargaro')]], 'Circuit Of The Americas': ['Estados Unidos da America', '5513', [('Red Bull KTM Factory Racing', 25, 'Brad Binder'), ('Monster Energy Yamaha MotoGP', 20, 'Fabio Quartararo'), ('LCR Honda IDEMITSU', 16, 'Tetsuta Nagashima'), ('Tech3 KTM Factory Racing', 13, 'Remy Gardner'), ('Team SUZUKI ECSTAR', 11, 'Joan Mir'), ('WithU Yamaha RNF MotoGP Team', 10, 'Darryn Binder'), ('LCR Honda CASTROL', 9, 'Alex Marquez'), ('Repsol Honda Team', 8, 'Pol Espargaro'), ('Team SUZUKI ECSTAR', 7, 'Danilo Petrucci'), ('LCR Honda IDEMITSU', 6, 'Takaaki Nakagami'), ('Gresini Racing MotoGP', 5, 'Fabio Di Giannantonio'), ('Team SUZUKI ECSTAR', 4, 'Alex Rins'), ('Aprilia Racing', 3, 'Maverick Vitalis'), ('Aprilia Racing', 2, 'Lorenzo Savadori'), ('Prima Pramac Racing', 1, 'Johann Zarco'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Repsol Honda Team', 0, 'Marc Marquez'), ('Ducati Lenovo Team', 0, 'Jack Miller'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe')]], 'Termas de Rio Hondo': ['Argentina', '4806', [('Team SUZUKI ECSTAR', 25, 'Alex Rins'), ('Repsol Honda Team', 20, 'Pol Espargaro'), ('Repsol Honda Team', 16, 'Marc Marquez'), ('Prima Pramac Racing', 13, 'Johann Zarco'), ('Aprilia Racing', 11, 'Aleix Espargaro'), ('Monster Energy Yamaha MotoGP', 10, 'Franco Morbidelli'), ('Gresini Racing MotoGP', 9, 'Enea Bastianini'), ('Ducati Lenovo Team', 8, 'Francesco Bagnaia'), ('Aprilia Racing', 7, 'Lorenzo Savadori'), ('Monster Energy Yamaha MotoGP', 6, 'Fabio Quartararo'), ('Team SUZUKI ECSTAR', 5, 'Kazuki Watanabe'), ('Team SUZUKI ECSTAR', 4, 'Danilo Petrucci'), ('Ducati Lenovo Team', 3, 'Jack Miller'), ('Aruba.it Racing', 2, 'Michele Pirro'), ('LCR Honda IDEMITSU', 1, 'Tetsuta Nagashima'), ('Team SUZUKI ECSTAR', 0, 'Joan Mir'), ('Repsol Honda Team', 0, 'Stefan Bradl'), ('Mooney VR46 Racing Team', 0, 'Luca Marini'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Aprilia Racing', 0, 'Maverick Vitalis'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez')]], 'Pertamina Mandalika Circuit': ['Indonesia', '4301', [('Repsol Honda Team', 25, 'Pol Espargaro'), ('WithU Yamaha RNF MotoGP Team', 20, 'Cal Crutchlow'), ('Aprilia Racing', 16, 'Maverick Vitalis'), ('WithU Yamaha RNF MotoGP Team', 13, 'Andrea Dovizioso'), ('Gresini Racing MotoGP', 11, 'Enea Bastianini'), ('Team SUZUKI ECSTAR', 10, 'Kazuki Watanabe'), ('LCR Honda IDEMITSU', 9, 'Tetsuta Nagashima'), ('Team SUZUKI ECSTAR', 8, 'Danilo Petrucci'), ('Monster Energy Yamaha MotoGP', 7, 'Franco Morbidelli'), ('Repsol Honda Team', 6, 'Marc Marquez'), ('Team SUZUKI ECSTAR', 5, 'Joan Mir'), ('Repsol Honda Team', 4, 'Stefan Bradl'), ('Mooney VR46 Racing Team', 3, 'Luca Marini'), ('Ducati Lenovo Team', 2, 'Jack Miller'), ('Red Bull KTM Factory Racing', 1, 'Brad Binder'), ('Monster Energy Yamaha MotoGP', 0, 'Fabio Quartararo'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('Gresini Racing MotoGP', 0, 'Fabio Di Giannantonio'), ('Aprilia Racing', 0, 'Aleix Espargaro'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Ducati Lenovo Team', 0, 'Francesco Bagnaia'), ('Aprilia Racing', 0, 'Lorenzo Savadori'), ('Team SUZUKI ECSTAR', 0, 'Alex Rins'), ('LCR Honda IDEMITSU', 0, 'Takaaki Nakagami'), ('LCR Honda CASTROL', 0, 'Alex Marquez'), ('Mooney VR46 Racing Team', 0, 'Marco Bezzecchi')]], 'Lusail International Circuit': ['Qatar', '5380', [('Aprilia Racing', 25, 'Lorenzo Savadori'), ('LCR Honda IDEMITSU', 20, 'Takaaki Nakagami'), ('LCR Honda CASTROL', 16, 'Alex Marquez'), ('Ducati Lenovo Team', 13, 'Jack Miller'), ('Aprilia Racing', 11, 'Maverick Vitalis'), ('Repsol Honda Team', 10, 'Stefan Bradl'), ('Team SUZUKI ECSTAR', 9, 'Alex Rins'), ('Gresini Racing MotoGP', 8, 'Fabio Di Giannantonio'), ('Aprilia Racing', 7, 'Aleix Espargaro'), ('Mooney VR46 Racing Team', 6, 'Luca Marini'), ('Mooney VR46 Racing Team', 5, 'Marco Bezzecchi'), ('Ducati Lenovo Team', 4, 'Francesco Bagnaia'), ('Monster Energy Yamaha MotoGP', 3, 'Fabio Quartararo'), ('Repsol Honda Team', 2, 'Marc Marquez'), ('Team SUZUKI ECSTAR', 1, 'Joan Mir'), ('Tech3 KTM Factory Racing', 0, 'Remy Gardner'), ('Repsol Honda Team', 0, 'Pol Espargaro'), ('Prima Pramac Racing', 0, 'Jorge Martin'), ('Red Bull KTM Factory Racing', 0, 'Brad Binder'), ('Team SUZUKI ECSTAR', 0, 'Kazuki Watanabe'), ('Gresini Racing MotoGP', 0, 'Enea Bastianini'), ('WithU Yamaha RNF MotoGP Team', 0, 'Cal Crutchlow'), ('Monster Energy Yamaha MotoGP', 0, 'Franco Morbidelli'), ('Team SUZUKI ECSTAR', 0, 'Danilo Petrucci'), ('Tech3 KTM Factory Racing', 0, 'Raul Fernandez'), ('WithU Yamaha RNF MotoGP Team', 0, 'Darryn Binder'), ('Red Bull KTM Factory Racing', 0, 'Miguel Oliveira'), ('Aruba.it Racing', 0, 'Michele Pirro'), ('Prima Pramac Racing', 0, 'Johann Zarco'), ('WithU Yamaha RNF MotoGP Team', 0, 'Andrea Dovizioso'), ('LCR Honda IDEMITSU', 0, 'Tetsuta Nagashima')]]}

#Este código dá-nos a classificação por circuito.

def classificaoPorCircuito(circuitos, circuito):
    equipas = {}
    pilotos = {}

    prova = circuitos[circuito]
    classificacao = prova[2] #recebe a lista com equipa, pontos, piloto
    for c in classificacao:
        equipa, pontos, piloto = c
        if equipa in equipas:#se já existe em equipas
            equipas[equipa] = equipas[equipa] + pontos #soma os pontos
        else: # se não existe em equipas
            equipas[equipa] = pontos #atribui os pontos
        if piloto in pilotos:#se já existe em pilotos
            pilotos[piloto] = pilotos[piloto] + pontos
        else:
            pilotos[piloto] = pontos            
    #print(sorted(equipas.items(), reverse=True, key=lambda item: item[1]))
    #print(sorted(pilotos.items(), reverse=True, key=lambda item: item[1]))
    return sorted(equipas.items(), reverse=True, key=lambda item: item[1]), sorted(pilotos.items(), reverse=True, key=lambda item: item[1])

equipas_usa, pilotos_usa = classificaoPorCircuito(circuitos2022, 'Circuit Of The Americas')
print(equipas_usa)
print('\n')
print(pilotos_usa)
print('\n')

#O código infra dá-nos a classificação geral,
#permitindo escolher quantos queremos apresentar

def classificao(circuitos):
    equipas = {}
    pilotos = {}

    for x in circuitos:
        circuito = circuitos[x]
        classificacao = circuito[2]
        for c in classificacao:
            equipa, pontos, piloto = c
            if equipa in equipas:
                equipas[equipa] = equipas[equipa] + pontos
            else:
                equipas[equipa] = pontos
            if piloto in pilotos:
                pilotos[piloto] = pilotos[piloto] + pontos
            else:
                pilotos[piloto] = pontos            
    # print(sorted(equipas.items(), reverse=True, key=lambda item: item[1]))
    # print(sorted(pilotos.items(), reverse=True, key=lambda item: item[1]))
    return sorted(equipas.items(), reverse=True, key=lambda item: item[1]), sorted(pilotos.items(), reverse=True, key=lambda item: item[1])

def primeiros(equipas, limite):
    i = 0
    for e in equipas:
        if i < limite:
            print('{} {}'.format(e[0], e[1]))
        i = i+1

equipas2022, pilots2022 = classificao(circuitos2022)
# print(equipas2022)
# print(pilots2022)

primeiros(equipas2022, 10)
print('\n')
primeiros(pilots2022, 10)

```
